```python
import re
import pandas as pd
```

```python
filename = "./data/InstitutionCampus.csv"
df = pd.read_csv(filename)
```

```python
df.head()
```

```python
df.replace(to_replace="[^A-Za-z0-9 ]+", value="", regex=True, inplace=True)
```

```python
no_parent_filter = df.ParentName == "-"
df.loc[no_parent_filter, "ParentName"] = df.loc[no_parent_filter, "LocationName"]
```

```python
universities_n_address = df.LocationName + " " + df.Address
universities = df.ParentName

d = {
    "University": universities,
    "Features": universities_n_address
}

export_df = pd.DataFrame(d)
print(export_df.shape)
```

```python
export_df.to_csv("./data/academics.csv", index=False)
```
